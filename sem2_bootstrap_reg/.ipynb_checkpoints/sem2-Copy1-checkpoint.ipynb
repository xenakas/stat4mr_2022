```python
import numpy as np
import numpy.random as npr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_points = 25
x = np.linspace(1,10, n_points)
b1 = 3
b2 = 10
u = npr.normal(0,x**(1.5), len(x))
y = b1+b2*x + u
```

```python
import statsmodels.formula.api as sm
```

```python
df = pd.DataFrame({'x':x, 'y':y})
```

```python
olsmodel = sm.ols(formula='y~x', data=df)
```

```python
res = olsmodel.fit()
```

```python
res.params
```

```python
ypred = res.predict(df['x'])
```

```python
plt.scatter(x,y)
plt.plot(x,ypred)
```

```python
b1_boot = [] 
b2_boot = []

n_points = 25
n_boot = 100 

plt.figure()

for i in range(n_boot):
    sample_df = df.sample(n=n_points, replace = True)
    ols_boot = sm.ols(formula='y~x', data=sample_df)
    res_boot = ols_boot.fit()
    
    b1_boot.append(res_boot.params[0]) 
    b2_boot.append(res_boot.params[1]) 
    
    ypred_boot = res_boot.predict(sample_df['x'])
    plt.plot(sample_df['x'], ypred_boot, color='grey', alpha=0.2)

plt.scatter(x,y)
plt.plot(x,ypred)    
    
```

```python
sns.distplot(b2_boot)
```

### Bootstrapping Residuals

```python
df = pd.DataFrame({'x':x, 'y':y})
olsmodel = sm.ols(formula='y~x', data=df)
res = olsmodel.fit()
res.params
```

```python
ypred = res.predict(df['x'])
residuals = res.resid
```

```python
plt.scatter(x,y)
plt.plot(x,ypred)
```

```python
b1_bootres = []
b2_bootres = []

n_points = 25
n_boot = 100

for i in range(n_boot):
    boot_res = npr.choice(residuals, n_points, replace=True)
    y_boot = ypred + boot_res
    
    sample_df=pd.DataFrame({'x':x, 'y':y_boot})
    
    ols_boot = sm.ols(formula='y~x', data=sample_df)
    res_boot = ols_boot.fit()
    
    b1_bootres.append(res_boot.params[0]) 
    b2_bootres.append(res_boot.params[1]) 

    
    
```

```python
np.mean(b1_boot), np.mean(b1_bootres)
```

```python
sns.distplot(b1_boot)
sns.distplot(b1_bootres)
```

```python
np.mean(b2_boot), np.mean(b2_bootres)
```

```python
np.percentile(b2_boot,[2.5,97.5])
```

```python
np.percentile(b2_bootres,[2.5,97.5])
```

```python
sns.distplot(b2_boot)
sns.distplot(b2_bootres)
```

```python
b1_bootres2 = []
b2_bootres2 = []

n_points = 25
n_boot = 100

h_weights = np.diag(np.matmul(np.matmul(X, np.linalg.inv(np.matmul(X.T,X))), X.T))
a = (residuals-np.mean(residuals))/np.std(residuals)   

for i in range(n_boot):
    boot_res = npr.choice(residuals, n_points, replace=True)
        
    t = npr.choice(a, n_points, replace=True)
    
    y_boot = ypred + boot_res * t/(np.sqrt(1-h_weights))
    
    
    sample_df=pd.DataFrame({'x':x, 'y':y_boot})
    
    ols_boot = sm.ols(formula='y~x', data=sample_df)
    res_boot = ols_boot.fit()
    
    b1_bootres2.append(res_boot.params[0]) 
    b2_bootres2.append(res_boot.params[1]) 

    
    
```

```python
# sns.distplot(b1_boot)
sns.distplot(b1_bootres)
sns.distplot(b1_bootres2)
```
